### Vprašanja

#### 1. Regresija (100%) 
Za vsakega uporabnika postavite regresijski model. Uporabite eno ali več metod za učenje regresijskih modelov (linearna regresija, Ridge, Lasso, itd.).

Za vsakega od $n$ uporabnikov izberite ustrezni stolpec v matriki podatkov. Za uporabnika $i$ imamo torej:

* Vektor odziva $y^{(i)}$,
* Matriko podatkov $X^{(i)}$, ki vsebuje vse stolpce *razen* $i$.
    
Za lažjo predstavo si oglej zgornji tabeli. Nekajkrat (npr., trikrat) ponovite postopek preverjanja s pomočjo učne in testne množice:


* Množico filmov, ki si jih je uporabnik ogledal, *naključno* razdelite v razmerju 75% (učna množica) in 25% (testna množica).
* Naučite regresijski model na učni množici (izberite ustrezne vrstice v $X$ in $y$).
* Ovrednotite model na testni množici (ponovno izberite ustrezne vrstice v $X$ in $y$).

Oceno vrednotenja nato delite s številom poizkusov, da dobite končno oceno.

Poročajte o uspešnosti vašega modela. Pri tem se osredotočite na naslednja vprašanja:
* Utemeljite ustrezno mero vrednotenja. Ali model dobro napoveduje ocene?
* Z izbrano mero ocenite modele za vseh $n$ uporabnikov.

Kodo za odgovore lahko razdelite v več celic.

```python
# Predpriprava

import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, explained_variance_score
from sklearn.model_selection import train_test_split
import pandas as pd

# Read datasets
movies = pd.read_csv('podatki/ml-latest-small/movies.csv')
ratings = pd.read_csv('podatki/ml-latest-small/ratings.csv')

# Create a dict of unique movie ids and user ids
movie_ratings_map = dict.fromkeys(ratings['movieId'].unique())
user_ratings_map = dict.fromkeys(ratings['userId'].unique())

# Get number of ratings for each movie
for k in movie_ratings_map.keys():
    movie_ratings_map[k] = ratings[ratings['movieId'] == k]['rating'].count()
# Get number of ratings for each user
for k in user_ratings_map.keys():
    user_ratings_map[k] = ratings[ratings['userId'] == k]['rating'].count()
    
# Extract top 100 movies with over 100 ratings
top_100_movies = {k : v for k, v in movie_ratings_map.items() if v > 100}
top_movie_ids = list(sorted(top_100_movies, key=top_100_movies.get, reverse=True)[:100])
    
# Extract top 100 users with over 100 ratings
top_100_users = {k : v for k, v in user_ratings_map.items() if v > 100}
top_user_ids = list(sorted(top_100_users, key=top_100_users.get, reverse=True)[:100])

# Create matrix of data
dmatrix = dict.fromkeys(top_movie_ids)

for k in dmatrix.keys():
    dmatrix[k] = dict.fromkeys(top_user_ids, 0.0)
    
for k in dmatrix.keys():
    for l in dmatrix[k].keys():
        if ratings[(ratings['movieId'] == k) & (ratings['userId'] == l)]['rating'].count() > 0:
            dmatrix[k][l] = ratings[(ratings['movieId'] == k) & (ratings['userId'] == l)]['rating'].iloc[0]

# Flatten internal dicts
fmatrix = dict()
for k, v in dmatrix.items():
    fmatrix[k] = np.array(v.values())
    
# Convert dict to pandas dataframe for easier computation and column selection
dfmatrix = pd.DataFrame(dmatrix).T
print(dfmatrix.head())
```

```python
# Regresija

# Create dicts that will hold the results of the regression for each user
lasso = dict.fromkeys(top_user_ids)
ridge = dict.fromkeys(top_user_ids)
linear = dict.fromkeys(top_user_ids)

# How many times to predict
predict_counter = 3

# Loop through users
for k in lasso.keys():
    lasso[k] = [0, 0]
    ridge[k] = [0, 0]
    linear[k] = [0, 0]
    
    # Create models for each user
    mlasso = Lasso(alpha=0.1)
    mridge = Ridge(alpha=0.1)
    mlinear = LinearRegression()
    
    only_rated = dfmatrix[dfmatrix[k] > 0]
    
    # Label (Y)
    response_vector = only_rated[k]
    # Attributes (X)
    data_matrix = only_rated.drop(k, axis=1)
    
    # Predict multiple times
    for i in range(predict_counter):
        # Split data into training and test sets
        X_train, X_test, y_train, y_test = train_test_split(data_matrix, response_vector, test_size=0.25)
        
        # Train models
        mlasso.fit(X_train, y_train)
        mridge.fit(X_train, y_train)
        mlinear.fit(X_train, y_train)
        
        # Predict values
        hlx = mlasso.predict(X_test)
        hrx = mridge.predict(X_test)
        hix = mlinear.predict(X_test)
        
        exp_var_l = explained_variance_score(y_test, hlx)
        exp_var_r = explained_variance_score(y_test, hrx)
        exp_var_i = explained_variance_score(y_test, hix)
        
        # Evaluate models
        lasso[k][0] += mean_squared_error(y_test, hlx)
        lasso[k][1] += exp_var_l
        ridge[k][0] += mean_squared_error(y_test, hrx)
        ridge[k][1] += exp_var_r
        linear[k][0] += mean_squared_error(y_test, hix)
        linear[k][1] += exp_var_i
        
    # Divide by num of predictions to get real results
    lasso[k] = tuple(x / predict_counter for x in lasso[k])   
    ridge[k] = tuple(x / predict_counter for x in ridge[k])
    linear[k] = tuple(x / predict_counter for x in linear[k]) 

# Convert dicts to pandas dataframe for easier analysis of statistics/results    
dflasso = pd.DataFrame.from_dict(lasso, orient='index', columns=['MSE', 'Explained Variance'])
dfridge = pd.DataFrame.from_dict(ridge, orient='index', columns=['MSE', 'Explained Variance'])
dflinear = pd.DataFrame.from_dict(linear, orient='index', columns=['MSE', 'Explained Variance'])
print(dflasso.describe(), '\n', dfridge.describe(), '\n', dflinear.describe())  
```

Naloge sem se lotil tako, da sem najprej podatke združil v matriko, opisano v razdelku Predpriprava podatkov. Nato sem deklariral 3 slovarje, enega za Lasso regresijski model, enega za Ridge regresijski model, in enega za linearno regresijo. Nato sem šel v zanki čez vse uporabnike v matriki, in za vsakega 3x izračunal povprečno kvadratno napako ter pojasnjeno varianco.

Za glavno mero vrednotenja sem izbral povprečno kvadratno napako, saj dobro prikaže odstopanje od želene vrednosti in se mi zdi za uporabljene modele najbolj primerna.

Ugotovil sem, da najboljše rezultate predikcij vrača Lasso model, saj je bila njegova povprečna MSE enaka 0.820, kar je veliko bolje kot pri Ridge (1.200) in linearni regresiji (1.207). Prav tako so najmanjše (0.175) in največje (3.033) vrednosti MSE pri Lasso modelu dost manjše (boljše) kot pri Ridge (min: 0.139, max: 9.076) in linearni regresiji (min: 0.174, max: 9.450). Torej, če modele razvrstimo po metriki MSE, je najboljši model Lasso, drugi model Ridge, najslabši pa je model linearne regresije.

Predvidevam, da bi boljše rezultate lahko pridobil z večanjem št. uporabnikov (več podatkov, iz katerih se modeli učijo) in z povečanjem števila učenj modelov (namesto 3, naprimer 10).

```python
movies = pd.read_csv('podatki/ml-latest-small/movies.csv')

# Get titles of top 100 movies
titles = movies[movies['movieId'].isin(top_movie_ids)][['title', 'movieId']]

# Map movie ids to ratings
mvs = {
    260   : 4.1,
    527   : 4.3,
    1073  : 2.3,
    4993  : 4.5,
    593   : 4.4,
    316   : 2.1,
    364   : 2.0,
    6539  : 4.3,
    586   : 2.7,
    58559 : 3.1  
}

# Add new column of 0s
dfmatrix = dfmatrix.assign(me=0)

# Assign my ratings to the matrix
for k, v in mvs.items():
    dfmatrix.loc[k, 'me'] = v

lasso = Lasso(alpha=0.1)
ridge = Ridge(alpha=0.1)
linear = LinearRegression()

me_rated = dfmatrix[dfmatrix['me'] > 0]

# Label (Y)
response_vector = me_rated['me']
# Attributes (X)
data_matrix = me_rated.drop('me', axis=1)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data_matrix, response_vector, test_size=0.25)

# Train models
lasso.fit(X_train, y_train)
ridge.fit(X_train, y_train)
linear.fit(X_train, y_train)

tmp = dfmatrix[dfmatrix['me'] == 0].drop('me', axis=1)

# Predict values
hlx = lasso.predict(tmp)
hrx = ridge.predict(tmp)
hix = linear.predict(tmp)

# Print results and movie titles
for i in zip(tmp.index, hlx, hrx, hix):
    print('Movie: ', titles[titles['movieId'] == i[0]].values[0][0], '\n ' \
            'Lasso: {:0.2f}'.format(i[1]), '\n', \
                'Ridge: {:0.2f}'.format(i[2]), '\n', \
                    'Linear: {:0.2f}'.format(i[3]))
```

Če bi ocenil več filmov, predvidevam da bi bila odstopanja ocen dosti manjša.
